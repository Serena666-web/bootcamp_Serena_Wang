```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/wangyuhan/bootcamp_Serena_Wang/project/data/raw/api_yfinance_nvda_20250218-0815.csv")
df = df.rename(columns={'adj_close': 'nvda_price'})
df.head()
```

## TODO: Implement at least 2 engineered features here

```python

df['change in price'] = df['nvda_price'].diff().fillna(0) 

df['rolling_price'] = df['nvda_price'].rolling(20, min_periods=1).mean()
display(df)

plt.figure(figsize=(12,6))
plt.plot(df['date'], df['nvda_price'], label='NVDA Price', linewidth=1.5)
plt.plot(df['date'], df['rolling_price'], label='20-day SMA', linewidth=2)

plt.title("NVDA Price with 20-Day Rolling Average", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()
plt.grid(True)
plt.show()
```
